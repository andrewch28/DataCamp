# Introduction to Data Visualization with Seaborn

## Introduction to Seaborn

- Easy to use
- Works well with pandas data structures
- Built on top of matplotlib

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]

#!Notice that we have to specify x and y
sns.scatterplot(x=height, y=weight)
plt.figure(figsize=(4,3))
plt.show()
````

````python
gender = ["Female", "Female", "Female", "Female",
           "Male", "Male", "Male", "Male", "Male", "Male"]

sns.countplot(x=gender)
plt.show()
````

## Adding a third variable with hue

there is a special dataset tips which is included in seaborn by default

````python
tips = sns.load_dataset("tips")

hue_colors = {"Yes": "purple",
              "No": "orange"}

sns.scatterplot(x="total_bill", y="tip", data=tips,
                 hue="smoker", hue_order=["No", "Yes"],
                 palette = hue_colors)
plt.show()
````

## Introduction to relational plots and subplots

Relational plots ```relplot```:
- lineplot
- scatterplot

We use relplot cause it allows us to create subplots in a single figure

````python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            col="smoker" #could have used row
            )

plt.show()
````

If you have too many values (like several dayes of the week you can use ```col_wrap``` to wrap it into 2 lines instead of 1 and the use```col_order``` to specify the order)

````python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            col="day", #could have used row
            col_wrap=2,
            col_order=["Thur",
                       "Fri",
                       "Sat",
                       "Sun"])

plt.show()
````

## Customizing scatter plots

size and hue

````python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            size="size",
            hue="size"
)

plt.show()
````

style

````python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            hue="smoker",
            style="smoker"
)

plt.show()
````

alpha

````python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            alpha=0.4
)

plt.show()
````

## Introduction to line plots

````python
mpg = pd.read_csv("databases/mpg.csv")

sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line") 
plt.show()
````

````python
sns.relplot(x="model_year", y="horsepower", 
            data=mpg,
            kind="line", 
            errorbar=None,
            style="origin", 
            hue="origin",
            markers=True,
            dashes=False)
plt.show()
````

## Categorical plots (count plots, bar plots)

- Involve a categorical variable (a couple of categories)
- Comparisons between groups

- we use ```catplot()```
- parameters ```col``` and ```row``` work the same way

````python
sns.set_context("paper")

category_order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"]
survey = pd.read_csv("databases/student-alcohol-consumption.csv")
sns.catplot(x="study_time", 
            data=survey,
            kind="count",
            order = category_order)
plt.show()
````

The difference of bar plots is that they display mean of quantative variable per category and no the counts of it

````python
sns.set_context("talk")

sns.catplot(x="day",
            y="total_bill",
            data=tips,
            kind="bar",
            errorbar=None,
            hue="day")

plt.show()
````

## Box plots

- shows the distribution of quantitative data
- see the median, spread, skewness and outliers
- facilitates comparisons between groups

````python
sns.set_context("paper")

order = ["Dinner", "Lunch"]

g = sns.catplot(x="time",
                y="total_bill",
                data=tips,
                kind="box"
                ) #to ommit outliers
plt.show()
````
